## BFI

```python
from calinet.core.pheno import score_bfi
df = score_bfi("Z:\\CALINET2\\converted\\bologna\\phenotype\\bfi30_english.tsv")
df.head()
```

```python
df_wide = df.loc[:, df.columns.str.contains("_total")].copy()
df_long = df_wide.melt(
    var_name="Trait",
    value_name="Score"
)
df_long["Trait"] = df_long["Trait"].str.replace("_total", "", regex=False)
df_long.head()
```

```python
pl = plotting.LazyBar(
    df_long,
    x="Trait",
    y="Score",
    hue="Trait",
    figsize=(3.54,3.54),
    fontname="Arial",
    sns_offset=4,
    y_label="score",
    add_labels=True,
    sns_rot=35,
    fancy=True,
    cmap="Set1"
)
```
